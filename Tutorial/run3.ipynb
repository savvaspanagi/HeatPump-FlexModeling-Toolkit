```python
import pandas as pd
import numpy as np
import os
np.random.seed(42)
%load_ext autoreload
%autoreload 2

# === Paths ===
path_in = os.path.join("../Data")  # parent directory
file_path = os.path.join(path_in, "dataset4.xlsx")
df = pd.read_excel(file_path)

# Rename columns
rename_mapping = {
        'T_ext': 'OUTDOOR_TEMP',
        'heating_total [kW]': 'PHI_H',
        'GHI': 'PYRANOMETER',
        'T_int': 'INDOOR_TEMP'
    }
df = df.rename(columns=rename_mapping)
df['PHI_H']=df['PHI_H']*1000
# Convert 't' to datetime and set as index

df['Time'] = pd.to_datetime(df['Time'])
df = df.set_index('Time')
df = df.resample("30min").mean()
surface_area = 2100  # mÂ² 
```

```python
df['OUTDOOR_TEMP']=df['OUTDOOR_TEMP'] + 273
df['INDOOR_TEMP']=df['INDOOR_TEMP']+273
```

```python
split_index = int(0.8 * 0.4*len(df))
end_index = int(0.4*len(df)) 
train_df = df.iloc[:split_index]
val_df = df.iloc[split_index:end_index]
```

```python
from models import *
from simulate import *
from plot import *

# --- Bounds ---
bounds_3r2c = {
    "R_in_e": (0.00018, 0.102),
    "R_e_a": (0.0003, 0.170),
    "R_in_a": (0.048, 0.272),
    "C_in": (360e3, 2040e3),
    "C_e": (360e3, 2040e3),
    "Ain": (0, 0),
    "Ae": (0, 0),
    "fh": (1,1)
}
param_defaults = {"R_in_e": 0.06, "R_e_a": 0.10, "R_in_a": 0.160, "C_in": 1200e3, "C_e": 1200e3, "Ain": 0, "Ae": 0,  "fh": 1}

# --- Build and Solve Model ---
model, parameters = build_greybox_model(
    train_df=train_df,
    model_type="3R2C",
    bounds=bounds_3r2c,
    show_plot=True,
    initialization=param_defaults
    ,deltaT=1
)
print(parameters['Objective'])
```

```python
parameters
```

```python
last_index = max(model.T_in.index_set())

# --- Validation Initialization (from last training value) ---
initialize_validation = {"Tin": model.T_in[last_index](), "Te": model.T_e[last_index]()}

# --- Run Simulation and Validation ---
validation, t_pred, t_true = simulate_and_evaluate(
    val_df=val_df,
    param_dict=parameters,
    model_type="3R2C",
    initializations=initialize_validation,
    show_plot=True
)

# --- Results ---
print(f"Validation RMSE: {validation['RMSE']:.4f}")

# --- Plot Residuals ---
plot_residual(t_pred=t_pred, t_true=t_true)
```
